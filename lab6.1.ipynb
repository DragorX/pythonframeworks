3. Find books that published by publisher with 0 or 1 vowel in its name.

```python
df_task3 = df[df["Publisher"].str.count("[AEIOU]*[aeiou]")<=1]
df_task3
```

4. Find authors whose last name starts and ends with consonants.

```python
df_task4 = df[df["Author"].str.extract('(\w+)')[0].fillna("").str.match("^[^AEIOU]*[^aeiou]$")]
df_task4
```

5. Find books with the maximum number of vowels in its title.

```python
df['MaxTitleVowels'] = df['Title'].str.count("[AEIOUaeiou]")
df.sort_values('MaxTitleVowels').tail(1)
```

7. Find author with 3 consonant in row in first name.

```python
df_task7 = df[df["Author"].str.extract("(\w+)(, )(\w+)(\w+)")[2].fillna("").str.contains("[^AEIOUaeiou]{3}")]
df_task7
```
